More efficient approach

```python
# Assume X is the grayscale image stored in a NumPy array
# If X is not loaded, load it using: X = cv.imread("a.png", cv.IMREAD_GRAYSCALE)

thresholds = [50, 70, 100, 150]
binarized_images = {}

# Loop through each threshold and apply binarization using NumPy operations
for t in thresholds:
    Z = np.zeros_like(X)  # Create an array of zeros with the same shape as X
    Z[X > t] = 255  # Set pixels greater than the threshold to 255 (white)
    binarized_images[t] = Z  # Store in dictionary

# Display all binarized images
plt.figure(figsize=(10, 5))

for i, t in enumerate(thresholds):
    plt.subplot(1, 4, i+1)
    plt.title(f"Threshold {t}")
    plt.imshow(binarized_images[t], cmap='gray')
    plt.axis('off')

plt.show()
```

Using openCV

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Assume X is the grayscale image loaded using OpenCV
# If X is not defined, load it using: X = cv.imread("a.png", cv.IMREAD_GRAYSCALE)

# Apply thresholding at different levels
_, Z50 = cv.threshold(X, 50, 255, cv.THRESH_BINARY)
_, Z70 = cv.threshold(X, 70, 255, cv.THRESH_BINARY)
_, Z100 = cv.threshold(X, 100, 255, cv.THRESH_BINARY)
_, Z150 = cv.threshold(X, 150, 255, cv.THRESH_BINARY)

# Display all binarized images
plt.figure(figsize=(10, 5))

thresholds = [50, 70, 100, 150]
images = [Z50, Z70, Z100, Z150]

for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.title(f"Threshold {thresholds[i]}")
    plt.imshow(images[i], cmap='gray')
    plt.axis('off')

plt.show()
```

Consider the color image stored in a.png.  Create a filter of [[-1,-1,-1][0,0,0][1,1,1]], and 
multiply this filter to each pixel value in the image. Display the image after filtering. 

What is Filtering?
Filtering is a technique used in image processing to enhance, modify, or extract features from an image. It involves applying a kernel (filter matrix) to an image to perform operations like edge detection, blurring, sharpening, etc. The filtering process works by performing a convolution between the kernel and local pixel neighborhoods.

Filter Explanation
The given filter (kernel) is:
        -1      -1      -1
         0       0       0
         1       1       1

This filter detects horizontal edges by subtracting the top row from the bottom row.
Areas with strong horizontal edges will have high intensity differences.

```python
# Load color image
img = cv.imread("a.png")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display

# Define the filter (3x3 kernel)
kernel = np.array([[-1, -1, -1], [ 0,  0,  0], [ 1,  1,  1]])

# Apply filter using convolution
filtered_img = cv.filter2D(img, -1, kernel)

# Display original and filtered image
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Filtered Image")
plt.imshow(filtered_img)
plt.axis("off")

plt.show()
```
