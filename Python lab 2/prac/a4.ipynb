Write a program to load a .csv file as a NumPy 1-D array. Find the maximum and minimum 
elements in the array. 
Hint: For the data, use the .csv file “Book1.csv” 

```python
data1 = np.loadtxt("book1.csv", skiprows=1, delimiter="\t", usecols=1, dtype=int)
print(f"Maximum is {max(data1)}, Minimum is {min(data1)}")
```

break down of above code:
"book1.csv" → The file name from which data is read.
skiprows=1 → Skips the first row (useful if the first row contains headers).
delimiter="\t" → Specifies tab (\t) as the column separator (TSV file).
If it's a comma-separated file (CSV), use delimiter=",".
usecols=1 → Extracts only the second column (indexing starts from 0).
dtype=int → Converts values to integers.

For the Numpy 1-D array as obtained in Q.1, sort the elements in ascending order. 

```python
data1 = np.sort(data1)
print(data1)
```

For the sorted Numpy 1-D array as obtained in Q.2, reverse the array and print.  

```python
data1 = data1[: :-1]
print(data1)
```
